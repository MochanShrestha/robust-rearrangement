```python
import cProfile
import pstats

# Perform random actions
env.reset()
n_steps = 5

def run_steps():
    for i in tqdm(range(n_steps)):
        action = np.random.uniform(-1, 1, size=(env.num_envs, 8))

        # Zero out all but the first 3 values
        action[:, 3:] = 0
        action[:, 6] = 1
        # Keep the gripper open
        action[:, 7] = -1
        env.step(action)

pr = cProfile.Profile()

start = datetime.now()
pr.enable()
run_steps()
pr.disable()
elapsed = datetime.now() - start

time_per_step = elapsed.total_seconds() / n_steps
env_steps_per_second = 1 / time_per_step * num_envs

print(f"Envs: {num_envs}, Time per step: {time_per_step:.2f}s, Env steps per second: {env_steps_per_second:.2f}")

ps = pstats.Stats(pr)
ps.sort_stats('cumulative')
ps.print_stats()
```
